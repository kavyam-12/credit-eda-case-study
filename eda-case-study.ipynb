```python
#find the percentage of missing values for all the columns
100*application_data.isnull().mean()
```

```python
#dropping the missing values where percentage is greater than 45.
cut_off=45
cols_to_drop=list(application_data.columns[100*application_data.isnull().mean()>cut_off])
print(cols_to_drop)
print(len(cols_to_drop))
```

```python
#dropping the missing values where percentage is greater than 45.
application_data=application_data.drop(cols_to_drop,axis=1)
```

```python
#shape after dropping the columns
application_data.shape
```

```python
100*application_data.isnull().mean()
```

```python
#Segregating categorical and numerical columns
categorical_columns=list(application_data.dtypes[application_data.dtypes=='object'].index)
numerical_columns=list(application_data.dtypes[application_data.dtypes!='object'].index)
```

```python
#Filling missing values of categorial columns
for col in numerical_columns:
    med=application_data[col].median()
    application_data[col]=application_data[col].fillna(med)
```

```python
#Filling missing values of numerical columns
for col in categorical_columns:
    mod=application_data[col].mode()[0]
    application_data[col]=application_data[col].fillna(mod)
```
